```r
my_data <- read.csv("Retail_data_CSV.csv")
```

```r
print("My Retail Data:")
print(my_data)
my_data$Total_Revenue <- my_data$Quantity * my_data$Price

```

```r
barplot(my_data$Total_Revenue,
        names.arg = my_data$Country,
        col = "steelblue",
        main = "Total Revenue by Country",
        xlab = "Country",
        ylab = "Revenue ($)")
```

```r
model <- lm(Quantity ~ Price, data = my_data)
my_data$Prediction_Error <- residuals(model)
hist(my_data$Prediction_Error,
     col="orange",
     main="Model Accuracy Check",
     xlab="Error Amount")
```
